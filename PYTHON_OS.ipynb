Creación y eliminación de directorios

```python
os.mkdir("Nueva_Carpeta_de_Prueba")

print(os.listdir("."))
```

Eliminación de directorios

```python
os.rmdir("Nueva_Carpeta_de_Prueba")

print(os.listdir("."))
```
