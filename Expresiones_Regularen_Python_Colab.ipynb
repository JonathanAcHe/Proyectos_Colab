```python
import re # Importamos el modulo regex
```

```python
texto = "Aprendiendo sobre Expresiones Regularen en Python y Colab"

print(re.search("sobrse", texto))
print(re.search("sobre", texto))
```

```python
cadena = "Vamos a aprender expr regulares"

text_buscar = "expr"

if re.search(text_buscar, cadena) is not None:
  print("Encontrado :)")
else:
  print("No Encontrado :( ")
```

```python
cadena = "Vamos a aprender expr regulares"

text_buscar = "expr"

result_busqueda = re.search(text_buscar, cadena)

print(result_busqueda.start())
print(result_busqueda.end())
print(result_busqueda.span())
```

```python
cadena = "Vamos a aprender expr regulares con python y con colab."

text_buscar = "con"

print(re.findall(text_buscar, cadena)) # Busca y entrega resultado como una arreglo o lista
print(len(re.findall(text_buscar, cadena))) #Cuantas veces aparece la palabra
```
